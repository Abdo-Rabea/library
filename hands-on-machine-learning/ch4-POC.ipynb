```python
import numpy as np
# trying linear regression from scratch

# create liner data with some noise
np.random.seed(42) # for reproducibility
x = np.random.rand(100, 1) * 2 # return a 100x1 array of random numbers between 0 and 2
y = 4 + 3 * x + np.random.rand(100, 1) 
# print(x)
X_b = np.c_[np.ones((100, 1)), x] # add x0 = 1 to each instance (m*n)
# print(np.ones((100, 1)))

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best) # should be close to [4, 3]
y_pred = X_b.dot(theta_best) # equals theta_best[0] + theta_best[1] * x
```

```python
# plot x and y
import matplotlib.pyplot as plt
plt.scatter(x,y)
# plot the line of best fit
plt.plot(x, 4 + 3 * x, color='red') 
plt.plot(x, y_pred, color='blue') 
plt.xlabel('x')
plt.ylabel('y')
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
## closed form using sklearn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
print(lin_reg.intercept_, lin_reg.coef_) # should be close to [4, 3] same as theta_best
```
